# This is Texa sharpshooter plot for 32 datasets. 
# The number is 32 for the reason that most algrothims did not report their training accuray.
# We find the training accuracy of 1-NN Euclidean Distance and BOSS at here: https://www2.informatik.hu-berlin.de/~schaefpa/boss/

# Search 'Texas '


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

def draw_Texa_sharpshooter(algrothim_train_acc,baseline_train_acc,algrothim_test_acc,baseline_test_acc,algrothim_name,baseline_name):
    fig = plt.figure()
    plt.xlabel('Expected gain')
    plt.ylabel('Actual Gain')
    plt.plot([1,1], [0,2], '--', linewidth=2, markersize=0,c ='C0')
    plt.plot([0,2], [1,1], '--', linewidth=2, markersize=0,c ='C0')
    plt.axis('equal')
    plt.scatter(algrothim_train_acc/baseline_train_acc,algrothim_test_acc/baseline_test_acc,
                label = algrothim_name+' VS '+baseline_name, 
                s =10,
                c ='C1')
    plt.legend(loc='lower left', bbox_to_anchor=(1, 0))
    plt.text(2, 2, 'TP', ha='left', fontsize=12)
    plt.text(0, 0, 'TN', ha='left', fontsize=12)
    plt.text(0, 2, 'FN', ha='left', fontsize=12)
    plt.text(2, 0, 'FP', ha='left', fontsize=12)

    
temp = np.asarray([[1.5,1.5],[0.5,0.5]])
result_path = './Results_on_32_datasets.csv'
file = pd.read_csv(result_path)
Euclidean_Train_Acc = file['1-NN Euclidean Distance Train Acc'].to_numpy()
Euclidean_Test_Acc = file['1-NN Euclidean Distance Test Acc'].to_numpy()

BOSS_Train_Acc = file['BOSS Train Acc'].to_numpy()
BOSS_Test_Acc = file['BOSS Test Acc'].to_numpy()

OS_Train_Acc = file['OS-CNN Train Acc'].to_numpy()
OS_Test_Acc = file['OS-CNN Test Acc'].to_numpy()


draw_Texa_sharpshooter(OS_Train_Acc, Euclidean_Train_Acc, OS_Test_Acc, Euclidean_Test_Acc,'OS','1-NN Euclidean Distance')
draw_Texa_sharpshooter(OS_Train_Acc, BOSS_Train_Acc, OS_Test_Acc, BOSS_Test_Acc,'OS','BOSS')
```
