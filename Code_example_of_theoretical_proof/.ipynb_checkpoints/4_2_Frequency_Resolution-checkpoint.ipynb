# The code shows the relationship between kernel size and frequency resolution

```python
import numpy as np
import matplotlib.pyplot as plt
Signal_length = 300
length_of_each_cycle = 50
time = np.arange(0, Signal_length*np.pi/length_of_each_cycle, np.pi/length_of_each_cycle);


Singal = np.sin(time)

plt.figure(figsize=(9, 3)) 
plt.subplot(1,2,1)
plt.plot(Singal)
plt.title('Singal')
plt.subplot(1,2,2)
plt.plot(abs(np.fft.fft(Singal))[0:int(Signal_length/2)])
plt.title('Singal in frequency space')
```

# with kernel size increase the pattern in frequency becomes sharper thus has better ability to select a specific frequency

```python

kernel_size_list = [5,10, 20, 30 , 50, 80]
for kernel_size in kernel_size_list:
    
    Kerenl = np.zeros((Signal_length))
    
    Kerenl[0:kernel_size] = Singal[0:kernel_size]
    
    plt.figure(figsize=(9, 3)) 
    plt.subplot(1,2,1)
    plt.plot(Kerenl)
    plt.title('Kerenl')
    plt.subplot(1,2,2)
    plt.plot(abs(np.fft.fft(Kerenl))[0:int(Signal_length/2)])
    plt.title('Kerenl in frequency space')
```
